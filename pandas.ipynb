# Метод 3 сигма. УДаление выбрасов

```python
import statistics
dataAge = data['Age']
mean = statistics.mean(dataAge)
stdev = statistics.stdev(dataAge)
lower_bound = mean - 3 * stdev
upper_bound = mean + 3 * stdev
print(lower_bound, upper_bound)
```

```python
data_cut = data.query('Age >=@lower_bound and Age < @upper_bound')
data_cut.describe()
```
